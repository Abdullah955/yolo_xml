```python
import xml.etree.cElementTree as ET
import os
import glob
import pandas as pd
import cv2
```

```python
# this function will read pictures in folder and return a dataframe of all picture attribute
def create_xml():
    
    files = []
    path_list = []
    height_list = []
    width_list = []
    channels_list = []
    
    directory = 'input'
    print('Progress ...')
    for filename in os.listdir(directory):
        if not filename.startswith('.'):
            path = os.path.abspath(filename) 
            files.append(filename)
            
            # [ Plate , random number , x1 , y1 , x2 , y2 , h , w ]
            # remmember that h and w are for the plate 
            # and we are going to use h and w for the full picture using cv2.shape
            
            
            imge_path = 'input/' + filename
            
            img = cv2.imread(imge_path,cv2.IMREAD_COLOR)
            height, width, channels = img.shape 
            
            
            imge_path = '/home/aalatir/Desktop/Projects/new_plate_detiction/input/' + filename
            path_list.append(imge_path)
            height_list.append(height)
            width_list.append(width)
            channels_list.append(channels)
            
            
            #print(height,width)
            
        df = pd.DataFrame({'file_name':files , 'full_path':path_list,'height':height_list\
                           ,'width':width_list,'channels':channels_list })
        
    print('DONE.')
    return df
files = create_xml()
```

```python

files["plate"] = files["file_name"].str.split("_", n = 8, expand = True)[0]
files["ID"] = files["file_name"].str.split("_", n = 8, expand = True)[1]
files["x1"] = files["file_name"].str.split("_", n = 8, expand = True)[2]
files["y1"] = files["file_name"].str.split("_", n = 8, expand = True)[3]
files["x2"] = files["file_name"].str.split("_", n = 8, expand = True)[4]
files["y2"] = files["file_name"].str.split("_", n = 8, expand = True)[5]
files["h"] = files["file_name"].str.split("_", n = 8, expand = True)[6]
files["w"] = files["file_name"].str.split("_", n = 8, expand = True)[7].str.split('.')[0][0]
files["extension"] = files["file_name"].str.split("_", n = 8, expand = True)[7].str.split('.')[0][1]
```

```python
# write the dataframe into xml file for yolo:
def write_xml(df):
    # read existing xml file and then modify it:
    root = ET.parse('Example.xml')
    print('Creating xmls ...')
    for i in df.index:

        root.find('.//folder').text  = 'Annotations'
        root.find('.//filename').text  = df.file_name[i]
        root.find('.//path').text  = df.full_path[i]
        root.find('.//width').text = df.width[i].astype(str)
        root.find('.//height').text  = df.height[i].astype(str)
        root.find('.//depth').text  = '3'
        root.find('.//name').text  = 'car_plate'
        root.find('.//xmin').text  = df.x1[i]#.astype(str)
        root.find('.//ymin').text  = df.y1[i]#.astype(str)
        root.find('.//xmax').text  = df.x2[i]#.astype(str)
        root.find('.//ymax').text  = df.y2[i]#.astype(str)
        file_name = 'Annotations/' +\
        str(df.file_name[i].split('.')[0]) + '.xml'
        
        root.write(file_name)
    print('DONE.')
write_xml(files)
```
